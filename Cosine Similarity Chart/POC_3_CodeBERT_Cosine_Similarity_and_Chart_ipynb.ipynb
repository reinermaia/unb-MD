```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo CSV atualizado
output_csv_path = "/content/sample_data/4_corrected_code_full_after_index.csv"

# Carregar o CSV com os resultados j√° calculados
df = pd.read_csv(output_csv_path)

# Gerar o histograma das similaridades
similarities = df['CodeBERT_Cosine Similarity'].dropna()  # Remove valores vazios
plt.hist(similarities, bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title("Distribution of Cosine Similarities")
plt.xlabel("Cosine Similarity")
plt.ylabel("Frequency")
plt.show()
```
